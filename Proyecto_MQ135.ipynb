```
# La información recoletada por el sensor MQ135 se encuentra recopilada en el repositorio
import pandas as pd
import math

m = pd.read_csv("https://raw.githubusercontent.com/MatthewGuz/PROYECTO/main/DATOS_ARDUINO_FINAL.csv", index_col=0)
m.info()
```

```
# Los dátos recopilados fueron: Óxidos de nitrógeno, Óxido de carbono y humo (mezcla de vapor de agua y ácido carbónico).
# Cada uno de estos gases se encuentran cuantificados en ppm o partículas por millón.


cantidad_de_datos = len(m)
print(f"En total se repilaron {cantidad_de_datos} datos")

print("")

promedio_NOX = m.NOX.mean()
print(f"El promedio de las mediciones de óxidos de nitrogeno fue de: {promedio_NOX}ppm")

promedio_CO = m.CO.mean()
print(f"El promedio de las mediciones de óxido de carbono fue de: {promedio_CO}ppm")

promedio_HUMO = m.HUMO.mean()
print(f"El promedio de las mediciones del humo fue de: {promedio_HUMO}ppm")

print("")

min_NOX = m.NOX.min()
print(f"El mínimo de las mediciones de óxidos de nitrogeno fue de: {min_NOX}ppm")

min_CO = m.CO.min()
print(f"El mínimo de las mediciones de óxido de carbono fue de: {min_CO}ppm")

min_HUMO = m.HUMO.min()
print(f"El mínimo de las mediciones del humo fue de: {min_HUMO}ppm")

print("")

max_NOX = m.NOX.max()
print(f"El máximo de las mediciones de óxidos de nitrogeno fue de: {max_NOX}ppm")

max_CO = m.CO.max()
print(f"El máximo de las mediciones de óxido de carbono fue de: {max_CO}ppm")

max_HUMO = m.HUMO.max()
print(f"El máximo de las mediciones del humo fue de: {max_HUMO}ppm")

print("")

me_18 = len(m[(m.DÍA == "18/03/2022")])
print(f"El día viernes 18 de marzo se realizaron {me_18} mediciones")

me_19 = len(m[(m.DÍA == "19/03/2022")])
print(f"El día sábado 19 de marzo se realizaron {me_19} mediciones")

me_20 = len(m[(m.DÍA == "20/03/2022")])
print(f"El día domingo 20 de marzo se realizaron {me_20} mediciones")

me_21 = len(m[(m.DÍA == "21/03/2022")])
print(f"El día lunes 21 de marzo se realizaron {me_21} mediciones")

me_22 = len(m[(m.DÍA == "22/03/2022")])
print(f"El día martes 22 de marzo se realizaron {me_22} mediciones")

me_23 = len(m[(m.DÍA == "23/03/2022")])
print(f"El día miércoles 23 de marzo se realizaron {me_23} mediciones")

me_24 = len(m[(m.DÍA == "24/03/2022")])
print(f"El día jueves 24 de marzo se realizaron {me_24} mediciones")
```
